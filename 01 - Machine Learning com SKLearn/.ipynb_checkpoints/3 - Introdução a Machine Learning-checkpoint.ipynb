```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/1b7d5475863c15f484ac495bd70975cf/raw/16aff7a0aee67e7c100a2a48b676a2d2d142f646/projects.csv"
dados = pd.read_csv(uri)
dados.head()
```

```python
a_renomear = {
    "unfinished":"nao_finalizado",
    "expected_hours":"horas_esperadas",
    "price":"preco"
}
dados = dados.rename(columns=a_renomear)
dados.head()
```

```python
trocar = {
    0:1,
    1:0
}
dados["finalizado"] = dados["nao_finalizado"].map(trocar)
dados.head()
```

```python
dados.tail()
```

```python
x = dados[["horas_esperadas", "preco"]]
y = dados["finalizado"]
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

SEED = 20
# Adicionando parmetro stratify no y para deixar as duas amostras com resultados proporcionais
treino_x, teste_x, treino_y, teste_y = train_test_split(x,y,
                                                        random_state=SEED,
                                                        test_size=0.25,
                                                        stratify=y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x),len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x,treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y,previsoes)
print("A acurcia foi %.2f%% " % (acuracia*100))
```

Repare que, aparentemente, estamos seguindo o mesmo padro de treino e teste. Ento o que j de errado com o nosso teste?

Quando estamos trabalhando com duas dimenses (e somente nesses casos), podemos colorir todos os pixels do grfico de acordo com as possveis previses do classificador que criamos. Por exemplo, ser ou no vendido um projeto na faixa de valor 25000 com tempo de construo de 60h?

O algortimo deve ser capaz de aprender a curva ascendente que vemos no grfico: quanto mais bem pago e com maior tempo de elaborao, maior chance do projeto ser de fato concludo.

Existem diversas maneiras de realizarmos esse procedimento, mas so maneiras de trabalhar com matrizes e que no necessariamente sero executadas explicitamente no nosso dia-a-dia. Como estamos em um caso bastante especfico, apenas com duas dimenses, iremos abord-las, j que nos ajudaro a compreender com clareza qual  o problema do nosso algortimo.

Do conjunto teste_x, coletaremos o valor mnimo (min()) para horas_esperadas, e atribuiremos o nome x_min a esse valor. Faremos um procedimento similar para x_max, mas dessa vez coletando o valor mximo. Repetiremos essa estrutura para teste_y, mas agora trabalhando com preco mnimo e mximo. Por fim, imprimiremos o resultado.

```python
x_min = teste_x.horas_esperadas.min()
x_max = teste_x.horas_esperadas.max()
y_min = teste_x.preco.min()
y_max = teste_x.preco.max()
print(x_min, x_max,y_min,y_max)
```

O x est indo de 1 a 100, enquanto y varia de 101 a 27738. Agora definiremos quantos pixels haver em cada dimenso; se multiplicamos 100 x 100, isto , 100 pixels para cada dimenso (altura e largura), teremos um total de 10.000 pontos, o que  o suficiente.

Entre x_min e x_max, cada espao ter o tamanho mximo do grfico dividido pelo total de pixels ((x_max - x_min)/pixels.

Por fim, adicionaremos o mtodo arange() da biblioteca do Numpy.

```python
import numpy as np

pixels = 100
np.arange(x_min, x_max, (x_max - x_min)/pixels)
```

A partir de 1, foram realizadas somas de 0.99 at chegar ao nmero99.01, dividindo proporcionalmente o espao do eixo_x no grfico. Faremos o mesmo procedimento para o eixo_y, aproveitando para nomear cada uma dessas operaes:

```python
pixels = 100
eixo_x = np.arange(x_min, x_max, (x_max - x_min)/ pixels)
eixo_y = np.arange(y_min, y_max, (y_max - y_min)/ pixels)
```

Agora a ideia  criarmos um grid entre os dois eixos, multiplicando as opes de acordo com as diferentes possibilidade de custo de projeto e tempo de execuo. Temos na biblioteca do Numpy o np.meshgrid(), que consegue mesclar um grid . Esse mtodo recebe como parmetros o eixo_x e eixo_y, e devolve o contedo  xx e yy. Vamos analisar xx:

```python
xx, yy = np.meshgrid(eixo_x, eixo_y)
xx
```

Com o cdigo estruturado desta maneira, o que temos  a repetio do eixo x 100 vezes, e o mesmo ocorre com eixo y. Ou seja, ainda no mesclamos realmente esses dados. Para fazermos isso, usaremos o xx.ravel():

```python
xx, yy = np.meshgrid(eixo_x, eixo_y)
xx.ravel()
```

Faremos o mesmo procedimento para yy, e ento concatenaremos xx e yy por meio de np.c_. Essa operao nos devolver os pontos.

Ao executarmos o cdigo veremos a concatenao, com os devidos pares em cada eixo.

```python
xx, yy = np.meshgrid(eixo_x, eixo_y)
pontos = np.c_[xx.ravel(), yy.ravel()]
pontos
```

Agora, nosso trabalho , a partir do modelo, fazer as previses (predict() para todos esses pontos. Como resultado, teremos uma srie de nmeros, que so as classificaes. Chamaremos essas classificaes de Z.

```python
Z = modelo.predict(pontos)
```

Precisamos nos atentar para o formato (shape) do array, que  10000, mas se analisarmos xx, verificaremos que suas dimenses so 100, 100:

```python
Z = modelo.predict(pontos)
Z.shape
```

```python
xx.shape
```

Precisamos redimensionar (reshape) o array de 10000 de acordo com xx.

```python
Z = modelo.predict(pontos)
Z = Z.reshape(xx.shape)
Z
```

Finalmente, poderemos plotar esses dados. Para isso, usaremos a biblioteca Matplotlib, cujas importaes costumam ser plt. Novamente, queremos que os pontos sejam espalhados (scatter()), mas no usaremos o scatterplot() do Seaborn pois, nesse caso, queremos um controle mais refinado dessa plotagem.

O mtodo scatter() dever receber teste_x.horas_esperadas e teste_x.preco. Em seguida, definiremos a cor por meio do argumentoc= recebendo teste_y.

```python
import matplotlib.pyplot as plt

plt.scatter(teste_x.horas_esperadas, teste_x.preco, c=teste_y)
```

Para reduzirmos o tamanho dos pontos e melhorarmos a visualizao dos dados, usaremos o parmetro s=1 (de size):

```python
plt.scatter(teste_x.horas_esperadas, teste_x.preco, c=teste_y, s=1)
```

J temos o plot bsico, nossa prxima tarefa  desenhar a classificao que estipulamos: se o valor for 0, pintaremos de uma cor, se for 1, de outra. Alm disso, entre esses nmeros, adicionaremos uma cor diferente para que possamos visualizar a curva que separa as duas classificaes do algortimo.

Para tanto, usaremos plt.contourf() (de countor, ou "contorno"), que receber xx, yy e Z. Para que a tonalidade no seja muito forte, escreveremos alpha=0.3, aumentando a transparncia.

```python
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(teste_x.horas_esperadas, teste_x.preco, c=teste_y, s=1)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

SEED = 15
np.random.seed(SEED)
# Adicionando parmetro stratify no y para deixar as duas amostras com resultados proporcionais
treino_x, teste_x, treino_y, teste_y = train_test_split(x,y,                                                        
                                                        test_size=0.25,
                                                        stratify=y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x),len(teste_x)))

modelo = SVC(gamma="auto")
modelo.fit(treino_x,treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y,previsoes)
print("A acurcia foi %.2f%% " % (acuracia*100))



xx, yy = np.meshgrid(eixo_x, eixo_y)
pontos = np.c_[xx.ravel(), yy.ravel()]
Z = modelo.predict(pontos)
Z = Z.reshape(xx.shape)
# DECISION BOUNDARY
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(teste_x.horas_esperadas, teste_x.preco, c=teste_y, s=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
SEED = 15
np.random.seed(SEED)
# Adicionando parmetro stratify no y para deixar as duas amostras com resultados proporcionais
raw_treino_x, raw_teste_x, treino_y, teste_y = train_test_split(x,y,                                                        
                                                        test_size=0.25,
                                                        stratify=y)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x),len(teste_x)))

scaler = StandardScaler()
scaler.fit(raw_treino_x)
treino_x = scaler.transform(raw_treino_x)
teste_x = scaler.transform(raw_teste_x)

modelo = SVC(gamma="auto")
modelo.fit(treino_x,treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y,previsoes)
print("A acurcia foi %.2f%% " % (acuracia*100))

```

```python
data_x = teste_x[:,0]
data_y = teste_x[:,1]

x_min = data_x.min()
x_max = data_x.max()
y_min = data_y.min()
y_max = data_y.max()

pixels = 100
eixo_x = np.arange(x_min, x_max, (x_max - x_min)/ pixels)
eixo_y = np.arange(y_min, y_max, (y_max - y_min)/ pixels)

xx, yy = np.meshgrid(eixo_x, eixo_y)
pontos = np.c_[xx.ravel(), yy.ravel()]
Z = modelo.predict(pontos)
Z = Z.reshape(xx.shape)
# DECISION BOUNDARY
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(data_x, data_y, c=teste_y, s=1)
```
